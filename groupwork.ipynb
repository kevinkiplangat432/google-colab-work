```python

# Print students with at least one subject score above 90.
# Find student(s) with the lowest average.
# Print grades distribution in text form (e.g., A:2, B:1…)

students = ['Alice', 'Bob', 'Charlie', 'David', 'Emma', 'Fiona', 'George', 'Hannah', 'Ian', 'Julia', 'Kevin', 'Laura', 'Michael', 'Nina', 'Oliver']
math_scores = [78, 56, 90, 45, 70, 82, 67, 59, 95, 61, 88, 73, 50, 97, 66]
science_scores = [85, 64, 92, 52, 75, 80, 68, 55, 91, 60, 86, 77, 49, 99, 62]
english_scores = [82, 58, 88, 48, 72, 79, 65, 54, 93, 63, 84, 70, 51, 96, 69]


```

# *Zip the list together so that python knows that they are related*

```python
zipped = list(zip(students, math_scores, science_scores, english_scores ))
print(zipped)
```

**Algorithm for determining the student**


*   create an empty list to later store the students
*   loop through each item separately but in zipped.
*   check if the items in the different list satisfy the condition above 90
*   append the student whose score satisfied the condition to list if true.
*   print the list









```python
students_above90 = []
for n, m, s, e in zip(students, math_scores, science_scores, english_scores):

   if m>90 or s>90 or e  > 90:
    students_above90.append(n)

# def print_students(x):
#   for x in students_above90:
#     print(x)

# print_students(students_above90)

# alternatively
print( f"The students with a score of above ninety are {students_above90}" )
```

**Algorithm for computing average**

*   Use list comprehension and zip to find the averages of each student.
*   determine what each students average is about

*   use min to find the minimum value
*   find the index of that minimum value


*   find the coresponding index for the student
*   print / output







```python
 averages = [round((m+s+e)/3) for n, m, s, e in zip (students, math_scores, science_scores, english_scores)]
zip_avg = list(zip(students, averages))
min_value = min(averages)
min_index = averages.index(min_value)
student = students[min_index]

print(f"Lowest avearage is from {student} with an average of {min_value}" )
print (f"all average values from each student {zip_avg}")

```

**Grade distribution for all scores in general**

```python
merged_list = math_scores + science_scores + english_scores
a = 0
b = 0
c = 0
d = 0
e = 0
f = 0

for score in merged_list:
    if score >= 80:
        a += 1
    elif score >= 70:
        b += 1
    elif score >= 60:
        c += 1
    elif score >= 50:
        d += 1
    else:
        f += 1



print(merged_list)
print (f"A:{a}")
print (f"B:{b}")
print (f"C:{c}")
print (f"D:{d}")
print (f"F:{e}")

```
